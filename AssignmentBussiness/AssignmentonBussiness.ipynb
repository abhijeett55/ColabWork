**Face to Face Communications**

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def analyze_interference(file_path):
    # Load dataset
    if not os.path.exists(file_path):
        print(f"Error: File not found at {file_path}")
        return

    df = pd.read_csv(file_path)

    # Selecting relevant columns
    columns = [
        "1.On a Scale of 1 to 5, How satisfied are you with the balance between your screen time and face to face interactions?",

        "2. On a Scale of 1 to 5, How Confident are you in Maintaining face-to-face communication compared to previous digital communication?",

        "3. On a Scale of 1 to 5, How often do you feel compelled to check your device during face-to-face interactions, even when its Unnecessary?",

        "4.On a scale of 1 to 5, how much do you feel your screen time has reduced on a face-to-face interactions?",

        "5.On a scale 1 to 5 , How much guilt or regret do you feel after reading too much time on screen instead of interacting face-face with someone?"

    ]

    # Check if all columns exist in dataset
    missing_columns = [col for col in columns if col not in df.columns]
    if missing_columns:
        print(f"Error: Missing columns in the dataset: {missing_columns}")
        return

    df = df[columns]

    # Convert to numeric
    df = df.apply(pd.to_numeric, errors='coerce')

    # Descriptive Statistics
    print("Descriptive Statistics:")
    print(df.describe())

    # Correlation heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title("Correlation Heatmap of Survey Responses")
    plt.show()

    # Distribution plots
    df.hist(figsize=(12, 8), bins=5, edgecolor='black')
    plt.suptitle("Distribution of Survey Responses", fontsize=16)
    plt.show()

    # Box plots
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df)
    plt.xticks(rotation=45)
    plt.title("Box Plot of Survey Responses")
    plt.show()



    # Check if columns exist
    df = df[[col for col in columns if col in df.columns]]

    # Convert to numeric (handle errors)
    df = df.apply(pd.to_numeric, errors='coerce')

    # Drop NaN rows (optional)
    df.dropna(inplace=True)

    # Debugging: Check data types
    print(df.dtypes)

    # Check if df is empty after conversion
    if df.empty:
        print("Error: No numeric data available for plotting.")
        return

    # Histogram
    plt.figure(figsize=(12, 8))
    df.hist(figsize=(12, 8), bins=5, edgecolor='black')
    plt.suptitle("Distribution of Survey Responses", fontsize=16)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()





import os

# Example Usage:
file_path = r"/content/sample_data/Face-to-face communications (Responses) - Form responses 1.csv"
analyze_interference(file_path)

```
